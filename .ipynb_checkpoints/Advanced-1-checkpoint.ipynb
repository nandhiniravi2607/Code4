# We will now learn about Containers

# A Docker Container helps create reproducible packages that contain code and its execution environment. These packages can run on any machine and is independent of the configuration of the machine. 

# Docker container is required since Runner machines on Github have defined software and environment installed that may not be the best option for you for your application. So, you run your code in your defined environment in a docker container and that container is still hosted by a runner but you are not bound by the environment of the runner anymore!
